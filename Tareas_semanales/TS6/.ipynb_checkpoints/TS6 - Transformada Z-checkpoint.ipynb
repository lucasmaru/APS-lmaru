## Transformada Z y Respuesta en frecuencia
Dado las siguientes ecuaciones en diferencias de los siguientes sistemas, que representan un filtro de media móvil: 

a) $y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)$
 

b) $y(n)=x(n−4)+x(n−3)+x(n−2)+x(n−1)+x(n)$
 

c)  $y(n)=x(n)−x(n−1)$
 

d)  $y(n)=x(n)−x(n−2)$
 

Se pide:

1) Hallar $T(z)=\frac{Y(z)}{X(z)}$
.
2) Calcular su respuesta en frecuencia de módulo y fase.

3) Simular y validar la respuesta en frecuencia de todos los sistemas con Numpy.



### Punto 1 - Funciones de transferencia

Para cada sistema, se obtiene $ T(z) = \frac{Y(z)}{X(z)}$:

---

**a)**

Ecuación:  
$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n)
$$

Transformada Z:  
$$
Y(z) = X(z)\cdot z^{-3} + X(z)\cdot z^{-2} + X(z)\cdot z^{-1} + X(z)\cdot z^0
$$

Transferencia:  
$$
T(z) = \frac{Y(z)}{X(z)} = z^{-3} + z^{-2} + z^{-1} + 1 = \frac{z^3 + z^2 + z^1 + z^0}{z^3}  
$$

$$
\boxed{T(z) = \frac{z^3 + z^2 + z^1 + 1}{z^3}}
$$
---

**b)**  
Ecuación:  
$$
y(n)=x(n−4)+x(n−3)+x(n−2)+x(n−1)+x(n)
$$

Transformada Z:  
$$
Y(z) = X(z)\cdot z^{-4} + X(z)\cdot z^{-3} + X(z)\cdot z^{-2} + X(z)\cdot z^{-1} + X(z)\cdot z^0
$$

Transferencia:  
$$
T(z) = \frac{Y(z)}{X(z)} = z^{-4} + z^{-3} + z^{-2} + z^{-1} + 1 = \frac{z^4 + z^3 + z^2 + z^1 + z^0}{z^4}
$$

$$
\boxed{T(z) = \frac{z^4 + z^3 + z^2 + z^1 + 1}{z^4}}
$$
---

**c)**  
Ecuación:
$$
y(n)=x(n)−x(n−1)
$$

Transformada Z:  
$$
Y(z) = X(z)-X(z)\cdot z^{-1}
$$

Transferencia:  
$$
T(z) = \frac{Y(z)}{X(z)} = -z^{-1} + z^{0} = \frac{z^1 - z^0}{z^1}
$$

$$
\boxed{T(z) = \frac{z^1 - 1}{z^1}}
$$
---

**d)**  
Ecuación:
$$
y(n)=x(n)−x(n−2)
$$

Transformada Z:  
$$
Y(z) = X(z)-X(z)\cdot z^{-2}
$$

Transferencia:  
$$
T(z) = \frac{Y(z)}{X(z)} = -z^{-2} + z^{0} = \frac{z^2 - z^0}{z^2}
$$

$$
\boxed{T(z) = \frac{z^2 - 1}{z^2}}
$$
---

### Punto 2 - Respuesta en frecuencia

Para cada sistema, se remplaza $ \left. T(z) \right|_{z = e^{j\Omega}}$

---
**a)**  
Transferencia: 
$$
T(z) = \frac{z^3 + z^2 + z^1 + 1}{z^3}
$$

Evaluacion en circunferencia unitaria:
$$
T(e^{j\Omega}) = \frac{e^{3j\Omega} + e^{2j\Omega} + e^{j\Omega} + 1}{e^{3j\Omega}} = e^{-j\frac{3}{2}\Omega} \left( e^{j\frac{3}{2}\Omega} + e^{j\frac{1}{2}\Omega} + e^{-j\frac{1}{2}\Omega} + e^{-j\frac{3}{2}\Omega} \right) = \left[ 2\cos\left(\frac{3}{2}\Omega\right) + 2\cos\left(\frac{1}{2}\Omega\right) \right] e^{-j\frac{3}{2}\Omega}
$$

$$
\begin{array}{c}
\boxed{
\text{Módulo: }|T(e^{j\Omega})| = |2\cos\left(\frac{3}{2}\Omega\right) + 2\cos\left(\frac{1}{2}\Omega\right)| \\
 \\
\text{Fase: }\angle T(e^{j\Omega}) = -\frac{3}{2} \Omega
}
\end{array}
$$

---
**b)**  
Transferencia: 
$$
T(z) = \frac{z^4 + z^3 + z^2 + z^1 + 1}{z^4}
$$

Evaluación en circunferencia unitaria:
$$
T(e^{j\Omega}) = \frac{e^{4j\Omega} + e^{3j\Omega} + e^{2j\Omega} + e^{j\Omega} + 1}{e^{4j\Omega}} = e^{-2j\Omega} \left( e^{2j\Omega} + e^{j\Omega} + 1 + e^{-j\Omega} + e^{-2j\Omega} \right) = \left[ 2\cos(2\Omega) + 2\cos(\Omega) + 1 \right] \cdot e^{-2j\Omega}
$$


$$
\begin{array}{c}
\boxed{
\text{Módulo: }|T(e^{j\Omega})| = |2\cos\left(2\Omega\right) + 2\cos\left(\Omega\right) + 1| \\
 \\
\text{Fase: }\angle T(e^{j\Omega}) = -2 \Omega
}
\end{array}
$$

---
**c)**  
Transferencia: 
$$
T(z) = \frac{z^1 - 1}{z^1}
$$

Evaluación en circunferencia unitaria:
$$
T(e^{j\Omega}) = \frac{e^{j\Omega} - 1}{e^{j\Omega}} = e^{\frac{-1}{2}j\Omega} \left( e^{\frac{1}{2}j\Omega} - e^{\frac{-1}{2}j\Omega} \right) = \left[ 2j\sin(\frac{1}{2}\Omega)\right] \cdot e^{\frac{-1}{2}j\Omega}
$$

$$
\begin{array}{c}
\boxed{
\text{Módulo: }|T(e^{j\Omega})| = |2j\cdot\sin\left(\frac{1}{2}\Omega\right)| \\
 \\
\text{Fase: }\angle T(e^{j\Omega}) = \frac{-1}{2} \Omega
}
\end{array}
$$

---
**d)**  
Transferencia: 
$$
T(z) = \frac{z^2 - 1}{z^2}
$$

Evaluación en circunferencia unitaria:
$$
T(e^{j\Omega}) = \frac{e^{2j\Omega} - 1}{e^{2j\Omega}} = e^{-j\Omega} \left( e^{j\Omega} - e^{-j\Omega} \right) = \left[ 2j\cdot\sin(\Omega)\right] \cdot e^{-j\Omega}
$$

$$
\begin{array}{c}
\boxed{
\text{Módulo: }|T(e^{j\Omega})| = |2j\cdot\sin\left(\Omega\right)| \\
 \\
\text{Fase: }\angle T(e^{j\Omega}) = -\Omega
}
\end{array}
$$
