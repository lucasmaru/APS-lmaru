## Introducción teórica

La **Transformada Z** es una herramienta fundamental en el análisis de sistemas lineales e invariantes en el tiempo (LTI) en tiempo discreto. Dada una señal $ x(n) $, su transformada Z se define como:

$$
X(z) = \sum_{n=-\infty}^{\infty} x[n] z^{-n}, z= r\cdot e^{j\Omega}
$$

Esta transformada permite trabajar con sistemas en el dominio complejo \( z \), facilitando el estudio de estabilidad, respuesta en frecuencia, y diseño de filtros.

Una propiedad clave utilizada en este trabajo es la **propiedad del retardo temporal**, que establece que:

$$
\mathcal{Z}\{x[n - k]\} = z^{-k} X(z)
$$

Es decir, un **desplazamiento temporal** hacia la derecha (retardo) en la señal original se traduce en una **multiplicación por $ z^{-k} $** en el dominio Z.

Esta propiedad nos permite obtener la función de transferencia de un sistema directamente a partir de su ecuación en diferencias.

### Punto 3- Verificación con Numpy
Primero se elabora un código con ayuda de IA para armar los polinomios en el dominio Zeta y graficar tanto módulo como fase:

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue May 27 10:32:19 2025

@author: lmaru
"""

import numpy as np
import matplotlib.pyplot as plt

# Grilla de frecuencias normalizadas en [0, π]
omega = np.linspace(0, np.pi, 1024)

# Diccionario de coeficientes h[n] para cada inciso
sistemas = {
    'a': [1, 1, 1, 1],
    'b': [1, 1, 1, 1, 1],
    'c': [1, -1],
    'd': [1, 0, -1]
}

# Crear figuras
plt.figure(figsize=(12, 8))

# Armar la respuesta en frecuencia evaluando la suma: H(e^{jω}) = sum h[n] e^{-jωn}
for i, (clave, h) in enumerate(sistemas.items()):
    #Crea un array del mismo tamaño que omega para guardar la respuesta en frecuencia.
    H = np.zeros_like(omega, dtype=complex) 
    for n, hn in enumerate(h):
        H += hn * np.exp(-1j * omega * n)
    
    # Módulo y fase
    modulo = np.abs(H)
    fase = np.unwrap(np.angle(H))

    # Gráficos
    plt.subplot(4, 2, 2*i+1)
    plt.plot(omega, modulo)
    plt.title(f'Módulo $|T(e^{{jω}})|$ - Inciso {clave}')
    plt.xlabel('ω [rad/muestra]')
    plt.grid()

    plt.subplot(4, 2, 2*i+2)
    plt.plot(omega, fase)
    plt.title(f'Fase $∠T(e^{{jω}})$ - Inciso {clave}')
    plt.xlabel('ω [rad/muestra]')
    plt.grid()

plt.tight_layout()
plt.show()
```

## Discusión de resultados

### Observaciones sobre las discontinuidades y la fase

En los gráficos generados con NumPy se observan distintos comportamientos de la **fase** según el sistema analizado. En particular, se distinguen **dos tipos de variaciones bruscas**:

---

**1. Saltos de fase de $ \pi $** (incisos a y b):

En los incisos **a)** y **b)**, se observan **saltos abruptos de $ \pi $** en la fase en las frecuencias donde el módulo de $ T(e^{j\Omega}) $ se anula.

Esto ocurre porque en esas frecuencias, el sistema tiene **ceros reales sobre la circunferencia unitaria**, lo cual implica un **cambio de signo** en la respuesta compleja $ H(e^{j\Omega}) $. Ese cambio genera un salto real de fase de $ \pi $, y no es un artefacto de visualización. Se puede observar claramente en el mapa de polos y zeros.

Por ejemplo:

- En el inciso **a)**, hay un cero en $ \Omega = \pi $
- En el inciso **b)**, hay dos ceros en $ \Omega = \frac{2\pi}{5} $ y $ \frac{4\pi}{5} $

En esos puntos aparecen los saltos de fase en los gráficos. Es importante destacar que la discontinuidad evitable no afecta la derivabilidad de la función y que la pendiente se mantiene consntante antes y despues de la discontinuidad.


**2. Desfase constante por términos imaginarios $ j\sin(\cdot) $ — incisos c y d:**

En los incisos **c)** y **d)**, se observa que la fase se acerca a $ \frac{\pi}{2} $ cuando $ \Omega \to 0 $, pero no debido a un salto o a una anulación del módulo, sino por una propiedad de la función compleja.

Es decir, **el seno aparece multiplicado por $ j $**, producto de la expansión de Euler de $ e^{-j\Omega}$ . Este término $ j\sin(\cdot) $ introduce un **desfase constante de $ \frac{\pi}{2} $** en la fase, que domina cerca de $ \Omega = 0 $.

Este fenómeno explica por qué, en los gráficos de fase, se observa una especie de upset de $ \frac{\pi}{2} $. Es solo un efecto que introduce la estructura compleja de la función. Es válido destacar que solo sucede en sistemas FIR donde se verifica la existencia de una antisimetría (asociado a los senos) y por tal motivo no vemos este efecto en los items a) y b).

---
### Análisis de ceros del módulo en el inciso b)

La función de transferencia del inciso b) es:

$$
T(z) = 1 + z^{-1} + z^{-2} + z^{-3} + z^{-4}
\Rightarrow H(e^{j\Omega}) = \sum_{n=0}^{4} e^{-j n \Omega}
$$

Esta expresión es una suma geométrica de razón $ r = e^{-j\Omega} $, por lo que se puede reescribir como:

$$
H(e^{j\Omega}) = \frac{1 - e^{-j5\Omega}}{1 - e^{-j\Omega}}
$$

El módulo de la respuesta en frecuencia es:

$$
|H(e^{j\Omega})| = \left| \frac{1 - e^{-j5\Omega}}{1 - e^{-j\Omega}} \right|
$$

Los ceros del módulo se encuentran donde el **numerador** se anula:

$$
1 - e^{-j5\Omega} = 0 \Rightarrow e^{-j5\Omega} = 1 \Rightarrow -5\Omega = 2\pi k \Rightarrow \Omega = \frac{2\pi k}{5}
$$

Para \( k = 1, 2, 3, 4 \), obtenemos los ceros en:

$$
\Omega = \frac{2\pi}{5}, \quad \frac{4\pi}{5}, \quad \frac{6\pi}{5}, \quad \frac{8\pi}{5}
$$

De estos, los que se encuentran dentro del intervalo $[0, \pi]$ son:

$$
\boxed{\Omega = \frac{2\pi}{5} \approx 1.26 \quad \text{y} \quad \Omega = \frac{4\pi}{5} \approx 2.51}
$$

Esto explica por qué el gráfico del módulo del inciso b) presenta **dos ceros dentro del rango [0, π] detalle que ignore en el gráfico a mano.**

En particular, incluso sin realizar cálculos numéricos, podemos anticipar varias propiedades de los filtros FIR:

- Cuando los coeficientes del filtro son **simétricos o antisimétricos**, la fase resultante es **lineal** en frecuencia.
- El **retardo de grupo** (o pendiente de la fase) es constante y está dado por:
  
  $$
  \tau_g = \frac{N - 1}{2}
  $$

  donde \( N \) es la cantidad de muestras del filtro. Este valor puede ser entero o fraccionario dependiendo de si \( N \) es impar o par, respectivamente.

- El **eje de simetría** del numerador coincide con el **retardo de grupo** del sistema.

Estas observaciones nos permiten **obtener de antemano** mucha información del sistema solo con observar la función transferencia.

No obstante, la exactitud, resolución y detalle que ofrecen las herramientas digitales son indispensables para:

- detectar ceros con precisión,
- observar fenómenos sutiles en la fase,
- y validar de manera cuantitativa las predicciones realizadas a mano.

En definitiva, la **complementariedad entre análisis manual y verificación computacional** demuestra ser clave para profundizar la comprensión de los sistemas en tiempo discreto.
