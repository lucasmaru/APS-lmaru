## Enunciado

En la TS7 se experimentaron técnicas de filtrado lineal para la mitigación de interferencias en la señal electrocardiográfica (ECG). Tal vez la interferencia que más exigió el diseño de los filtros digitales ha sido el **movimiento de línea de base del ECG**, el cual consiste en oscilaciones de muy baja frecuencia (< 0.5 Hz) y, en ocasiones, de alta energía.

En esta tarea semanal se analizarán técnicas **no lineales** de estimación y substracción de la señal de movimiento de base \( b[n] \), con el objetivo de recuperar la señal libre de esta interferencia:

$$
\hat{x}[n] = s[n] - \hat{b}[n]
$$

Siendo:

$$
s[n] \quad \text{la señal de ECG contaminada}
$$

$$
\hat{x}[n] \quad \text{la señal filtrada (estimación del ECG sin interferencias)}
$$

Se pide implementar ambas estimaciones de \( b[n] \) detalladas a continuación:

---

### 1) Filtro de Mediana

Se estima la señal $\hat{b}[n]$ a partir de la siguiente expresión anidada:

$$
\hat{b}[n] = \text{med}_{600} \left( \text{med}_{200}(s[n]) \right)
$$

Donde:

$$
\text{med}_T(\cdot)
$$

denota el operador de mediana aplicado sobre una ventana temporal de duración \( T \) milisegundos.

En forma expandida:

$$
m[n] = \text{mediana} \left( s[n], s[n-1], \dots, s[n - 200 \cdot f_s] \right)
$$

$$
\hat{b}[n] = \text{mediana} \left( m[n], m[n-1], \dots, m[n - 600 \cdot f_s] \right)
$$

**Ayuda:** Puede utilizar la implementación del filtro de mediana provista en `scipy.signal.medfilt`.

---

### 2) Interpolación mediante splines cúbicos

A partir de las detecciones provistas en `ecg.mat` (campo `qrs_detections`), puede estimarse el **nivel isoeléctrico**, es decir, el momento donde la actividad eléctrica cardíaca es aproximadamente nula.

El intervalo entre la onda **P** y **Q** es el momento más adecuado para estimar las características de la interferencia que origina el movimiento de la línea de base. En ese momento, el impulso cardíaco es demorado al atravesar el nódulo aurículo-ventricular.

Para localizar el segmento **PQ**, puede utilizar la localización de cada latido (líneas rojas verticales):

$$
n_i
$$

y anticipar un tiempo adecuado:

$$
n_0
$$

de forma tal de no incluir ni la onda **P** ni la **Q**.

Esto resultará en un conjunto de valores:

$$
S = \left\{ \left( m_i, s(m_i) \right) \ \big| \ i = 1, 2, \dots, Q \right\}
$$

donde:

$$
m_i = n_i - n_0
$$

y \( Q \) es la cantidad de latidos en el registro (1903).

Recuerde que la señal de línea de base debe estar muestreada a la misma frecuencia que el ECG, por lo que debería remuestrear dicha señal. Puede utilizar para ello una interpolación mediante splines cúbicos:

$$
\hat{b}[n] = \text{spline}_3(S, n) \quad \text{para } n = 1, 2, \dots, N
$$

donde la función:

$$
\text{spline}_3
$$

será la responsable de realizar la interpolación para las \( N \) muestras de la señal \( s \).

## 1) Filtro de mediana

- Se filtra la señal original en dos etapas de mediana:

  $$
  m[n] = \text{mediana}_{200}(s[n]), \quad \hat{b}[n] = \text{mediana}_{600}(m[n])
  $$

- Se calcula la señal estimada libre de movimiento de línea de base:

  $$
  \hat{x}[n] = s[n] - \hat{b}[n]
  $$

- Se grafica las tres señales:
  - La señal original \( s[n] \)
  - La componente de baja frecuencia estimada $ \hat{b}[n] $
  - La señal final filtrada $ \hat{x}[n] $



```python
#%%IMPORTACIONES
import numpy as np
from scipy import signal as sig

import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import plot_plantilla
import scipy.io as sio

#%%ECG
fs = 1000 # Hz, la frecuencia de muestreo a la que fue sampleada la señal del archivo
mat_struct = sio.loadmat('./ECG_TP4.mat') #Leo el archivo
ecg_one_lead = mat_struct['ecg_lead'].flatten() #Tomo la parte que quiero
N_ecg = len(ecg_one_lead)
latidos = mat_struct['qrs_detections'].flatten()

#%%FILTRADO DE MEDIANA
# Ventanas en milisegundos
ventana_1_ms = 200
ventana_2_ms = 600

# Conversión de ventanas a número impar de muestras
ventana_1 = int(ventana_1_ms * fs / 1000)
ventana_2 = int(ventana_2_ms * fs / 1000)

if ventana_1 % 2 == 0:
    ventana_1 += 1
if ventana_2 % 2 == 0:
    ventana_2 += 1

# Primer paso: filtro de mediana sobre la señal original
m = sig.medfilt(ecg_one_lead, kernel_size=ventana_1)

# Segundo paso: filtro de mediana sobre el resultado anterior
b_hat = sig.medfilt(m, kernel_size=ventana_2)

# Señal filtrada: ECG sin el movimiento de línea de base
ecg_filtrada = ecg_one_lead - b_hat

#%% PLOTEO RESULTADOS
GRAFICA_1=True
if GRAFICA_1:
    plt.figure(figsize=(12, 4))
    plt.plot(ecg_one_lead, label='Señal original',alpha=0.5)
    plt.plot(b_hat, label='Estimación b̂[n] (mov. línea base)', linewidth=1.5)
    plt.plot(ecg_filtrada, label='ECG filtrado (x̂[n])', linewidth=1, alpha=0.5)
    plt.xlim(0, fs*10)  # mostrar primeros 10 segundos
    plt.title('Filtro de Mediana Anidado')
    plt.xlabel('Muestra [n]')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
```

## 2) Interpolación mediante splines cúbicos


- Se extrae puntos isoeléctricos $ s[m_i] $ a partir de los latidos $ n_i $ desplazados por $ n_0 =100 ms$:

  $$
  m_i = n_i - n_0
  $$

- Se construye un conjunto de pares:

  $$
  S = \left\{ (m_i, s(m_i)) \right\}
  $$

- Se interpola esos valores usando un spline cúbico para estimar la línea de base:

  $$
  \hat{b}[n] = \text{spline}_3(S, n)
  $$

- Se calcula la señal filtrada:

  $$
  \hat{x}[n] = s[n] - \hat{b}[n]
  $$

- Se grafica la señal original \( s[n] \), la línea de base estimada $ \hat{b}[n] $, y la señal filtrada $\hat{x}[n]$.


```python
#%% INTERPOLACIÓN SPLINE CÚBICA

from scipy.interpolate import CubicSpline

# Definimos el desplazamiento n0 en milisegundos y lo convertimos a muestras
n0_ms = 100
n0 = int(n0_ms * fs / 1000)

# Generamos los mi = ni - n0, asegurándonos de no irnos fuera del rango
mi = latidos - n0
mi = mi[(mi >= 0) & (mi < N_ecg)]

# Creamos el conjunto de puntos S = {(mi, s[mi])}
si = ecg_one_lead[mi]

# Interpolamos con splines cúbicos
spline_interp = CubicSpline(mi, si)

# Evaluamos la interpolación en todos los puntos de la señal
b_hat_spline = spline_interp(np.arange(N_ecg))

# Señal filtrada
ecg_filtrada_spline = ecg_one_lead - b_hat_spline

#%% PLOTEO
GRAFICA_2=True
if GRAFICA_2:
    plt.figure(figsize=(12, 4))
    plt.plot(ecg_one_lead, label='Señal original')
    plt.plot(b_hat_spline, label='Estimación b̂[n] spline', linewidth=2)
    plt.plot(ecg_filtrada_spline, label='ECG filtrado spline', linewidth=1)
    plt.xlim(0, fs*10)
    plt.xlabel('Muestra [n]')
    plt.ylabel('Amplitud')
    plt.title('Interpolación del movimiento de línea de base mediante splines cúbicos')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
```

## 3) Filtro Adaptado (Matched Filter)

#### a) Explicación Conceptual

El **filtro adaptado** es una técnica de detección óptima cuando se conoce a priori la forma del pulso que se desea detectar en una señal ruidosa. En este caso, se parte de un patrón típico del complejo QRS (`qrs_pattern1`) como "forma base" del latido cardíaco.

---

#### 🧠 Principio del filtro adaptado

- El filtro se construye con la **respuesta impulsiva igual al patrón invertido temporalmente**.
- Al aplicarlo (por convolución o correlación), se **maximiza la salida** en aquellos puntos donde la señal contiene un pulso de forma similar al patrón.
- Este máximo ocurre cuando el patrón está **alineado** con un latido real, por lo tanto, la salida del filtro tiene **picos bien definidos** en esos instantes.

---

#### ⚙️ Aplicación en ECG

- Se toma el patrón de latido normal $h[n]$ y se calcula:

  $$
  y[n] = s[n] * h[-n] = \sum_k s[k] \cdot h[n + k]
  $$

- Esta operación equivale a una **correlación cruzada**, que se puede realizar directamente con `scipy.signal.correlate`.
- La señal $y[n]$ resultante tiene **picos positivos** en los momentos donde el ECG se asemeja al patrón.

---

#### ✅ Ventajas del enfoque

- Detección robusta frente al ruido, ya que se basa en la forma completa del pulso.
- No requiere segmentación previa, umbrales manuales ni modelado complejo.

#### ⚠️ Limitaciones

- La detección depende fuertemente de que el patrón coincida con los latidos del registro. Si la morfología del QRS cambia mucho (por ruido, fatiga o tipo de paciente), la detección se degrada.
- Puede haber **falsas detecciones** si hay segmentos de la señal que se parecen al patrón, pero no corresponden a latidos reales.

---

Esta técnica es especialmente útil en sistemas de detección automática donde se cuenta con un patrón representativo y se desea minimizar errores en presencia de ruido.


#### b) Implementación práctica


```python
#%% FILTRO ADAPTADO 

# Normalizamos la señal y el patrón
ecg_norm = ecg_one_lead / np.std(ecg_one_lead)
patron_qrs = mat_struct['qrs_pattern1'].flatten()
patron_qrs = patron_qrs / np.std(patron_qrs)

# Aplicamos correlación (matched filter)
y_matched = np.correlate(ecg_norm, patron_qrs, mode='same')
y_matched = y_matched / np.std(y_matched)

# Detectamos picos en la salida del filtro adaptado
umbral = 0.15  # relativo a std = 1
periodo_refractario = int(0.3 * fs)  # 300 ms
latidos_detectados, _ = sig.find_peaks(y_matched, height=umbral, distance=periodo_refractario)

#%% EVALUACIÓN DE DESEMPEÑO
tolerancia = int(0.05 * fs)  # ±50 ms
TP = 0
FP = 0
FN = 0
flag_detectado = np.zeros(len(latidos_detectados), dtype=bool)

for real in latidos:
    dentro = np.abs(latidos_detectados - real) <= tolerancia
    if np.any(dentro):
        TP += 1
        flag_detectado[np.where(dentro)[0][0]] = True
    else:
        FN += 1

FP = np.sum(~flag_detectado)

sensibilidad = TP / (TP + FN)
vpp = TP / (TP + FP)

print(f"TP = {TP}, FN = {FN}, FP = {FP}")
print(f"Sensibilidad = {sensibilidad:.3f}")
print(f"VPP = {vpp:.3f}")

#%% PLOTEO
t_ecg = np.arange(N_ecg) / fs

GRAFICA_3=True
if GRAFICA_3:
    plt.figure(figsize=(12, 4))
    plt.plot(t_ecg, y_matched, label='Salida del filtro adaptado', color='orange')
    plt.plot(t_ecg[latidos_detectados], y_matched[latidos_detectados], 'rx', label='Detecciones')
    plt.plot(t_ecg, ecg_norm, label='ECG normalizado', alpha=0.5)
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.title("Filtro Adaptado sobre el ECG")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
```

### Interpretación de Métricas del Detector por Filtro Adaptado

Se obtuvieron los siguientes resultados al comparar las detecciones con los latidos reales:

- **TP (verdaderos positivos)** = 1729  
- **FN (falsos negativos)** = 174  
- **FP (falsos positivos)** = 237  

A partir de estos valores se calcularon:

- **Sensibilidad**: 0.909  
  → El 90.9% de los latidos reales fueron correctamente detectados.

- **Valor Predictivo Positivo (VPP)**: 0.879  
  → El 87.9% de las detecciones realizadas corresponden a latidos reales.

Estas métricas indican un desempeño general muy bueno del filtro adaptado, aunque hay margen de mejora, especialmente en la reducción de **falsas alarmas (FP)** y la recuperación de **latidos no detectados (FN)**. Es probable que algunos errores se deban a latidos ectópicos que no coinciden bien con el patrón utilizado.


## Conclusión

En esta Tarea Semanal se implementaron y compararon distintas estrategias para la mitigación del movimiento de línea de base y la detección de latidos en señales de ECG, abordando tanto el filtrado como la identificación de eventos fisiológicos.

Se evaluaron tres enfoques:

1. **Filtro de mediana anidado**: permitió eliminar eficazmente las oscilaciones lentas de la señal sin afectar la morfología del complejo QRS. Esta técnica resultó particularmente útil para preservar las características de alta frecuencia del ECG, al evitar distorsiones introducidas por métodos más agresivos.

3. **Filtro adaptado**: aplicado como detector de latidos, demostró una alta sensibilidad y valor predictivo positivo al comparar la señal con un patrón típico del QRS. La evaluación con métricas objetivas mostró un buen rendimiento global, aunque con un número no despreciable de falsas alarmas, especialmente en presencia de latidos ventriculares cuya forma difiere del patrón utilizado.

Un aspecto destacable de esta tarea fue la **simplicidad de implementación** de las técnicas no lineales comparada con los métodos lineales trabajados previamente (especialmente los filtros FIR). Mientras que los filtros lineales requieren un diseño cuidadoso de la plantilla en frecuencia, elección del tipo de aproximación y ajustes finos de orden y estabilidad, los enfoques de esta TS se resolvieron con estructuras conceptuales claras, herramientas estándar (como `medfilt`, `CubicSpline` y `correlate`) y parámetros fácilmente interpretables en tiempo (milisegundos, ventana de tolerancia, etc.).

En conjunto, estas herramientas ilustran el valor del procesamiento digital de señales no solo para mejorar la calidad de registros biomédicos, sino también para permitir el desarrollo de detectores automáticos precisos, fundamentales en aplicaciones clínicas y de monitoreo continuo.

### 💎 Bonus: ¿Generalización del filtro adaptado?

Una pregunta relevante al trabajar con filtros adaptados es si el detector funcionaría igual de bien si se utilizara un patrón proveniente de **otro paciente** o de **otro registro**.

En principio, la efectividad del filtro adaptado depende de cuán representativo sea el patrón respecto de los latidos reales presentes en la señal. Dado que el complejo QRS puede variar considerablemente entre individuos (por diferencias anatómicas, posición de los electrodos, presencia de arritmias, etc.), un patrón extraído de otro paciente podría no alinearse correctamente con los latidos reales del nuevo registro. Esto daría lugar a una mayor cantidad de **falsos negativos** (latidos no detectados) y/o **falsos positivos** (detecciones espurias).

Por lo tanto, aunque el concepto del filtro adaptado es general, su **uso clínico confiable requiere patrones personalizados** o, al menos, patrones ajustados a características poblacionales comunes.
