En la TS7 se experimentaron t칠cnicas de filtrado lineal para la mitigaci칩n de interferencias en la se침al electrocardiogr치fica (ECG). En esta tarea semanal se analizar치n t칠cnicas **no lineales** de estimaci칩n y substracci칩n de la se침al de movimiento de base \( b[n] \), con el objetivo de recuperar la se침al libre de esta interferencia:

$$
\hat{x}[n] = s[n] - \hat{b}[n]

Siendo:

$$
s[n] \quad \text{la se침al de ECG contaminada}

$$
\hat{x}[n] \quad \text{la se침al filtrada (estimaci칩n del ECG sin interferencias)}

### 1) Filtro de Mediana

Se estima la se침al $\hat{b}[n]$ a partir de la siguiente expresi칩n anidada:

$$
\hat{b}[n] = \text{med}_{600} \left( \text{med}_{200}(s[n]) \right)
$$

Donde:

$$
\text{med}_T(\cdot)
$$

denota el operador de mediana aplicado sobre una ventana temporal de duraci칩n \( T \) milisegundos.

En forma expandida:

$$
m[n] = \text{mediana} \left( s[n], s[n-1], \dots, s[n - 200 \cdot f_s] \right)
$$

$$
\hat{b}[n] = \text{mediana} \left( m[n], m[n-1], \dots, m[n - 600 \cdot f_s] \right)
$$

### 2) Interpolaci칩n mediante splines c칰bicos

A partir de las detecciones provistas en `ecg.mat` (campo `qrs_detections`), puede estimarse el **nivel isoel칠ctrico**, es decir, el momento donde la actividad el칠ctrica card칤aca es aproximadamente nula.

El intervalo entre la onda **P** y **Q** es el momento m치s adecuado para estimar las caracter칤sticas de la interferencia que origina el movimiento de la l칤nea de base. En ese momento, el impulso card칤aco es demorado al atravesar el n칩dulo aur칤culo-ventricular.

Para localizar el segmento **PQ**, puede utilizar la localizaci칩n de cada latido (l칤neas rojas verticales):

$$
n_i
$$

y anticipar un tiempo adecuado:

$$
n_0
$$

de forma tal de no incluir ni la onda **P** ni la **Q**.

Esto resultar치 en un conjunto de valores:

$$
S = \left\{ \left( m_i, s(m_i) \right) \ \big| \ i = 1, 2, \dots, Q \right\}
$$

donde:

$$
m_i = n_i - n_0
$$

y \( Q \) es la cantidad de latidos en el registro (1903).

Recuerde que la se침al de l칤nea de base debe estar muestreada a la misma frecuencia que el ECG, por lo que deber칤a remuestrear dicha se침al. Puede utilizar para ello una interpolaci칩n mediante splines c칰bicos:

$$
\hat{b}[n] = \text{spline}_3(S, n) \quad \text{para } n = 1, 2, \dots, N
$$

donde la funci칩n:

$$
\text{spline}_3
$$

ser치 la responsable de realizar la interpolaci칩n para las \( N \) muestras de la se침al \( s \).

## 1) Filtro de mediana

- Se filtra la se침al original en dos etapas de mediana:

  $$
  m[n] = \text{mediana}_{200}(s[n]), \quad \hat{b}[n] = \text{mediana}_{600}(m[n])
  $$

- Se calcula la se침al estimada libre de movimiento de l칤nea de base:

  $$
  \hat{x}[n] = s[n] - \hat{b}[n]
  $$

- Se grafica las tres se침ales:
  - La se침al original \( s[n] \)
  - La componente de baja frecuencia estimada $ \hat{b}[n] $
  - La se침al final filtrada $ \hat{x}[n] $



```python
#%%IMPORTACIONES
import numpy as np
from scipy import signal as sig

import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import plot_plantilla
import scipy.io as sio

#%%ECG
fs = 1000 # Hz, la frecuencia de muestreo a la que fue sampleada la se침al del archivo
mat_struct = sio.loadmat('./ECG_TP4.mat') #Leo el archivo
ecg_one_lead = mat_struct['ecg_lead'].flatten() #Tomo la parte que quiero
N_ecg = len(ecg_one_lead)
latidos = mat_struct['qrs_detections'].flatten()

#%%FILTRADO DE MEDIANA
# Ventanas en milisegundos
ventana_1_ms = 200
ventana_2_ms = 600

# Conversi칩n de ventanas a n칰mero impar de muestras
ventana_1 = int(ventana_1_ms * fs / 1000)
ventana_2 = int(ventana_2_ms * fs / 1000)

if ventana_1 % 2 == 0:
    ventana_1 += 1
if ventana_2 % 2 == 0:
    ventana_2 += 1

# Primer paso: filtro de mediana sobre la se침al original
m = sig.medfilt(ecg_one_lead, kernel_size=ventana_1)

# Segundo paso: filtro de mediana sobre el resultado anterior
b_hat = sig.medfilt(m, kernel_size=ventana_2)

# Se침al filtrada: ECG sin el movimiento de l칤nea de base
ecg_filtrada = ecg_one_lead - b_hat

#%% PLOTEO RESULTADOS
GRAFICA_1=True
if GRAFICA_1:
    plt.figure(figsize=(12, 4))
    plt.plot(ecg_one_lead, label='Se침al original',alpha=0.5)
    plt.plot(b_hat, label='Estimaci칩n b팩[n] (mov. l칤nea base)', linewidth=1.5)
    plt.plot(ecg_filtrada, label='ECG filtrado (x팩[n])', linewidth=1, alpha=0.5)
    plt.xlim(0, fs*10)  # mostrar primeros 10 segundos
    plt.title('Filtro de Mediana Anidado')
    plt.xlabel('Muestra [n]')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
```

## 2) Interpolaci칩n mediante splines c칰bicos


- Se extrae puntos isoel칠ctricos $ s[m_i] $ a partir de los latidos $ n_i $ desplazados por $ n_0 =100 ms$:

  $$
  m_i = n_i - n_0
  $$

- Se construye un conjunto de pares:

  $$
  S = \left\{ (m_i, s(m_i)) \right\}
  $$

- Se interpola esos valores usando un spline c칰bico para estimar la l칤nea de base:

  $$
  \hat{b}[n] = \text{spline}_3(S, n)
  $$

- Se calcula la se침al filtrada:

  $$
  \hat{x}[n] = s[n] - \hat{b}[n]
  $$

- Se grafica la se침al original \( s[n] \), la l칤nea de base estimada $ \hat{b}[n] $, y la se침al filtrada $\hat{x}[n]$.


```python
#%% INTERPOLACI칍N SPLINE C칔BICA

from scipy.interpolate import CubicSpline

# Definimos el desplazamiento n0 en milisegundos y lo convertimos a muestras
n0_ms = 100
n0 = int(n0_ms * fs / 1000)

# Generamos los mi = ni - n0, asegur치ndonos de no irnos fuera del rango
mi = latidos - n0
mi = mi[(mi >= 0) & (mi < N_ecg)]

# Creamos el conjunto de puntos S = {(mi, s[mi])}
si = ecg_one_lead[mi]

# Interpolamos con splines c칰bicos
spline_interp = CubicSpline(mi, si)

# Evaluamos la interpolaci칩n en todos los puntos de la se침al
b_hat_spline = spline_interp(np.arange(N_ecg))

# Se침al filtrada
ecg_filtrada_spline = ecg_one_lead - b_hat_spline

#%% PLOTEO
GRAFICA_2=True
if GRAFICA_2:
    plt.figure(figsize=(12, 4))
    plt.plot(ecg_one_lead, label='Se침al original')
    plt.plot(b_hat_spline, label='Estimaci칩n b팩[n] spline', linewidth=2)
    plt.plot(ecg_filtrada_spline, label='ECG filtrado spline', linewidth=1)
    plt.xlim(0, fs*10)
    plt.xlabel('Muestra [n]')
    plt.ylabel('Amplitud')
    plt.title('Interpolaci칩n del movimiento de l칤nea de base mediante splines c칰bicos')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
```

## 3) Filtro Adaptado (Matched Filter)

#### a) Explicaci칩n Conceptual

El **filtro adaptado** es una t칠cnica de detecci칩n 칩ptima cuando se conoce a priori la forma del pulso que se desea detectar en una se침al ruidosa. En este caso, se parte de un patr칩n t칤pico del complejo QRS (`qrs_pattern1`) como "forma base" del latido card칤aco.

---

#### 游 Principio del filtro adaptado

- El filtro se construye con la **respuesta impulsiva igual al patr칩n invertido temporalmente**.
- Al aplicarlo (por convoluci칩n o correlaci칩n), se **maximiza la salida** en aquellos puntos donde la se침al contiene un pulso de forma similar al patr칩n.
- Este m치ximo ocurre cuando el patr칩n est치 **alineado** con un latido real, por lo tanto, la salida del filtro tiene **picos bien definidos** en esos instantes.

---

#### 丘뙖잺 Aplicaci칩n en ECG

- Se toma el patr칩n de latido normal $h[n]$ y se calcula:

  $$
  y[n] = s[n] * h[-n] = \sum_k s[k] \cdot h[n + k]
  $$

- Esta operaci칩n equivale a una **correlaci칩n cruzada**, que se puede realizar directamente con `scipy.signal.correlate`.
- La se침al $y[n]$ resultante tiene **picos positivos** en los momentos donde el ECG se asemeja al patr칩n.

---

#### 九 Ventajas del enfoque

- Detecci칩n robusta frente al ruido, ya que se basa en la forma completa del pulso.
- No requiere segmentaci칩n previa, umbrales manuales ni modelado complejo.

#### 丘멆잺 Limitaciones

- La detecci칩n depende fuertemente de que el patr칩n coincida con los latidos del registro. Si la morfolog칤a del QRS cambia mucho (por ruido, fatiga o tipo de paciente), la detecci칩n se degrada.
- Puede haber **falsas detecciones** si hay segmentos de la se침al que se parecen al patr칩n, pero no corresponden a latidos reales.

---

Esta t칠cnica es especialmente 칰til en sistemas de detecci칩n autom치tica donde se cuenta con un patr칩n representativo y se desea minimizar errores en presencia de ruido.


#### b) Implementaci칩n pr치ctica


```python
#%% FILTRO ADAPTADO 

# Normalizamos la se침al y el patr칩n
ecg_norm = ecg_one_lead / np.std(ecg_one_lead)
patron_qrs = mat_struct['qrs_pattern1'].flatten()
patron_qrs = patron_qrs / np.std(patron_qrs)

# Aplicamos correlaci칩n (matched filter)
y_matched = np.correlate(ecg_norm, patron_qrs, mode='same')
y_matched = y_matched / np.std(y_matched)

# Detectamos picos en la salida del filtro adaptado
umbral = 0.15  # relativo a std = 1
periodo_refractario = int(0.3 * fs)  # 300 ms
latidos_detectados, _ = sig.find_peaks(y_matched, height=umbral, distance=periodo_refractario)

#%% EVALUACI칍N DE DESEMPE칌O
tolerancia = int(0.05 * fs)  # 췀50 ms
TP = 0
FP = 0
FN = 0
flag_detectado = np.zeros(len(latidos_detectados), dtype=bool)

for real in latidos:
    dentro = np.abs(latidos_detectados - real) <= tolerancia
    if np.any(dentro):
        TP += 1
        flag_detectado[np.where(dentro)[0][0]] = True
    else:
        FN += 1

FP = np.sum(~flag_detectado)

sensibilidad = TP / (TP + FN)
vpp = TP / (TP + FP)

print(f"TP = {TP}, FN = {FN}, FP = {FP}")
print(f"Sensibilidad = {sensibilidad:.3f}")
print(f"VPP = {vpp:.3f}")

#%% PLOTEO
t_ecg = np.arange(N_ecg) / fs

GRAFICA_3=True
if GRAFICA_3:
    plt.figure(figsize=(12, 4))
    plt.plot(t_ecg, y_matched, label='Salida del filtro adaptado', color='orange')
    plt.plot(t_ecg[latidos_detectados], y_matched[latidos_detectados], 'rx', label='Detecciones')
    plt.plot(t_ecg, ecg_norm, label='ECG normalizado', alpha=0.5)
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.title("Filtro Adaptado sobre el ECG")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
```

### Interpretaci칩n de M칠tricas del Detector por Filtro Adaptado

Se obtuvieron los siguientes resultados al comparar las detecciones con los latidos reales:

- **TP (verdaderos positivos)** = 1729  
- **FN (falsos negativos)** = 174  
- **FP (falsos positivos)** = 237  

A partir de estos valores se calcularon:

- **Sensibilidad**: 0.909  
   El 90.9% de los latidos reales fueron correctamente detectados.

- **Valor Predictivo Positivo (VPP)**: 0.879  
   El 87.9% de las detecciones realizadas corresponden a latidos reales.

Estas m칠tricas indican un desempe침o general muy bueno del filtro adaptado, aunque hay margen de mejora, especialmente en la reducci칩n de **falsas alarmas (FP)** y la recuperaci칩n de **latidos no detectados (FN)**. Es probable que algunos errores se deban a latidos ect칩picos que no coinciden bien con el patr칩n utilizado.

### 游눑 Bonus: 쮾eneralizaci칩n del filtro adaptado?

Una pregunta relevante al trabajar con filtros adaptados es si el detector funcionar칤a igual de bien si se utilizara un patr칩n proveniente de **otro paciente** o de **otro registro**.

En principio, la efectividad del filtro adaptado depende de cu치n representativo sea el patr칩n respecto de los latidos reales presentes en la se침al. Dado que el complejo QRS puede variar considerablemente entre individuos (por diferencias anat칩micas, posici칩n de los electrodos, presencia de arritmias, etc.), un patr칩n extra칤do de otro paciente podr칤a no alinearse correctamente con los latidos reales del nuevo registro. Esto dar칤a lugar a una mayor cantidad de **falsos negativos** (latidos no detectados) y/o **falsos positivos** (detecciones espurias).

Por lo tanto, aunque el concepto del filtro adaptado es general, su **uso cl칤nico confiable requiere patrones personalizados** o, al menos, patrones ajustados a caracter칤sticas poblacionales comunes.
