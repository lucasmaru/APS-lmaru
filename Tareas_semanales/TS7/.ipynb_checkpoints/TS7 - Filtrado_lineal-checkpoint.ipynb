## a) Establecimiento de la plantilla de diseño

Para que la señal de ECG se asemeje a los latidos promedio, se requiere:
- **Suavidad de los trazos** → eliminar componentes de alta frecuencia (ruido).
- **Nivel isoeléctrico nulo** → eliminar componentes de muy baja frecuencia (desplazamientos de línea base o drift).

Por ello, se diseñará un **filtro pasa banda** digital con los siguientes requerimientos:

- **Banda pasante:** 1 Hz – 35 Hz
- **Atenuación fuera de banda:** 
  - < 0.1 Hz (rechazo de continua y drift)
  - > 50 Hz (rechazo de ruido muscular y componentes de alta frecuencia)
- **Ripple máximo en banda pasante:** 1 dB
- **Atenuación mínima en bandas de rechazo:** 40 dB

Esta plantilla se aplicará a los diferentes métodos de diseño de filtros digitales (IIR y FIR) que se evaluarán. 
A continuación el código utilizado apoyado en la librería pytc2:

```python
#%%IMPORTACIONES
import numpy as np
from scipy import signal as sig

import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import plot_plantilla
import scipy.io as sio

#%%ECG
fs_ecg = 1000 # Hz, la frecuencia de muestreo a la que fue sampleada la señal del archivo
mat_struct = sio.loadmat('./ECG_TP4.mat') #Leo el archivo
ecg_one_lead = mat_struct['ecg_lead'].flatten() #Tomo la parte que quiero
N_ecg = len(ecg_one_lead)

#%% PARÁMETROS - FILTRADO

# Frecuencia de muestreo
fs = 1000  # Hz
nyq = fs / 2  # Frecuencia de Nyquist

# Banda de atenuación (fuera de lo útil)
f_stop = np.array([.1, 50.])  # Hz

# Banda pasante deseada (contenido útil del ECG)
f_pass = np.array([1.0, 30.0])  # Hz

# Ripple permitido en la banda pasante
ripple_db = 1  # dB

# Atenuación mínima en la banda de stop
atten_db = 40  # dB

plt.figure(figsize=(10, 4))

"""Creamos un gráfico vacío donde se dibujará la plantilla la función se encarga de superponer 
zonas de paso y rechazo para un filtro determinado, como acá solo queremos la plantilla
vacía es necesario establecer manualmente los límites para que plt.axis() funcione
bien dentro de plot_plantilla."""

plt.axis([0, fs / 2, -atten_db - 10, ripple_db + 5])

plot_plantilla(filter_type='bandpass',
           fpass=f_pass,
           ripple=ripple_db,
           fstop=f_stop,
           attenuation=atten_db,
           fs=fs)

plt.title("Plantilla de diseño para el filtrado del ECG")
plt.tight_layout()
plt.show()
```

## b) Justificación de los valores de la plantilla

Para establecer los valores de la plantilla se consideraron tanto criterios fisiológicos como evidencia empírica obtenida a partir del análisis espectral de la señal ECG.

- **1–35 Hz (banda pasante):** la mayoría del contenido útil del ECG se encuentra en este rango. Las principales componentes morfológicas (ondas P, QRS y T) están contenidas en este espectro.  
  Además, en la Tarea Semanal 5 (TS5) se realizó una estimación espectral de la señal ECG mediante el método de Welch, observándose que aproximadamente el 98% de la energía total está contenida por debajo de los 30–35 Hz. Este análisis respalda cuantitativamente la elección de la banda pasante como representativa del contenido informativo del ECG.  

- **0–0.1 Hz y 50–500 Hz (bandas de rechazo):** el extremo inferior (por debajo de 0.5 Hz) se ataca para eliminar componentes de muy baja frecuencia, típicamente asociadas a interferencias por movimiento, desplazamiento de electrodos o fluctuaciones de línea base.  
  Por otro lado, el extremo superior (por encima de 50 Hz) busca suprimir interferencias de alta frecuencia como ruido electromagnético, incluyendo la línea eléctrica (50 Hz en Argentina), artefactos de electromiograma (EMG), y otras señales no deseadas.

- **Ripple permitido de 1 dB** en la banda pasante y una **atenuación mínima de 40 dB** en las bandas de rechazo fueron elegidos como compromisos razonables entre fidelidad en la banda útil y supresión efectiva del ruido, sin requerir órdenes de filtro excesivamente altos.


## c) Diseñe al menos dos filtros FIR y dos IIR para su comparación. Verifique que la respuesta en frecuencia responda a la plantilla de diseño.

Se diseñaron cuatro filtros distintos: dos de tipo FIR (ventana Hamming con `firwin2` y mínimos cuadrados con `firls`) y dos de tipo IIR (aproximaciones Butterworth y Chebyshev Tipo I).

Para verificar que los filtros cumplen con estos requisitos, se graficaron sus respuestas en frecuencia junto a la plantilla. Esto permite una inspección visual clara de si los filtros respetan los márgenes impuestos, tanto en ganancia como en las frecuencias de corte.


```python
#%% FUNCIÓN: GRAFICAR RESPUESTA DE UN FILTRO SOS
GRAFICAR = True
def graficar_filtro_sos(sos, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='Filtro'):
    w, hh = sig.sosfreqz(sos, worN=2048, fs=fs)
    orden = 2 * sos.shape[0]
    hh_dB = 20 * np.log10(np.maximum(np.abs(hh), 1e-10))
    fase = np.unwrap(np.angle(hh))
    demora = -np.diff(fase) / np.diff(w)
    w_med = (w[1:] + w[:-1]) / 2

    plt.figure(figsize=(10, 6))

    # Módulo
    plt.subplot(2, 1, 1)
    plt.plot(w, hh_dB, label=f'{etiqueta} (orden {orden})')
    plot_plantilla(filter_type='bandpass', fpass=f_pass, ripple=ripple_db,
                   fstop=f_stop, attenuation=atten_db, fs=fs)
    plt.ylabel("Ganancia [dB]")
    plt.xlabel("Frecuencia [Hz]")
    plt.title(f"Respuesta en módulo - {etiqueta}")
    plt.grid()
    plt.legend()

    # Fase y demora
    plt.subplot(2, 1, 2)
    plt.plot(w, fase, label='Fase')
    plt.plot(w_med, demora, label='Demora de grupo')
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Fase [rad] / Demora [s]")
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()    
#%% DISEÑO IIR - CHEBY2 + CAUER

# Diseño del filtro en formato SOS
mi_iir_1 = sig.iirdesign(wp=f_pass, ws=f_stop, gpass=ripple_db, gstop=atten_db,
                         analog=False, ftype='cheby2', output='sos', fs=fs)
mi_iir_2 = sig.iirdesign(wp=f_pass, ws=f_stop, gpass=ripple_db, gstop=atten_db,
                         analog=False, ftype='ellip', output='sos', fs=fs)

# Grilla para barrido en frecuencia
w, hh = sig.sosfreqz(mi_iir_1, worN=2048, fs=fs)

# Gráfico módulo y fase

if GRAFICAR:
    graficar_filtro_sos(mi_iir_1, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='IIR - Cheby2')
    graficar_filtro_sos(mi_iir_2, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='IIR - Cauer')

#%% FUNCIÓN: GRAFICAR RESPUESTA FIR
def graficar_fir(fir_coefs, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='FIR'):
    w_fir, hh_fir = sig.freqz(fir_coefs, worN=8000, fs=fs)
    hh_fir_dB = 20 * np.log10(np.maximum(np.abs(hh_fir), 1e-10))
    fase_fir = np.unwrap(np.angle(hh_fir))
    demora_fir = -np.diff(fase_fir) / np.diff(w_fir)
    w_fir_med = (w_fir[1:] + w_fir[:-1]) / 2

    plt.figure(figsize=(10, 6))

    # Módulo
    plt.subplot(2, 1, 1)
    plt.plot(w_fir, hh_fir_dB, label=f'{etiqueta} (orden {len(fir_coefs) - 1})')
    plot_plantilla(filter_type='bandpass', fpass=tuple(f_pass),
                   ripple=ripple_db, fstop=tuple(f_stop),
                   attenuation=atten_db, fs=fs)
    plt.ylabel("Ganancia [dB]")
    plt.xlabel("Frecuencia [Hz]")
    plt.title(f"Respuesta en módulo - {etiqueta}")
    plt.grid()
    plt.legend()

    # Fase y demora
    plt.subplot(2, 1, 2)
    plt.plot(w_fir, fase_fir, label='Fase')
    plt.plot(w_fir_med, demora_fir, label='Demora de grupo')
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Fase [rad] / Demora [s]")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

#%% DISEÑO FIR - FIRWIN2 Y FIRLS

# Orden común para ambos )
cant_coef_firwin2 = 22501 # Orden = cant_coef - 1
cant_coef_firls = 3001
orden_firwin2 = cant_coef_firwin2 - 1
orden_firls = cant_coef_firls - 1

# Grilla de diseño (Hz)
freq = [0, f_stop[0], f_pass[0], f_pass[1], f_stop[1], fs/2]
gain = [0, 0, 1, 1, 0, 0]

# FIR con firwin2 y ventana Hamming
mi_fir_1 = sig.firwin2(numtaps=cant_coef_firwin2, freq=freq, gain=gain, fs=fs, window='hamming')

# FIR con cuadrados mínimos (firls)
mi_fir_2 = sig.firls(numtaps=cant_coef_firls , bands=freq, desired=gain, fs=fs)



if GRAFICAR:
    graficar_fir(mi_fir_1, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='FIR - firwin2 (Hamming)')
    graficar_fir(mi_fir_2, fs, f_pass, f_stop, ripple_db, atten_db, etiqueta='FIR - firls')
```

## d) Filtre la señal y compare visualmente el resultado con la señal original. Describa qué observa.

Se aplicaron los filtros diseñados a la señal de ECG original utilizando la función `scipy.signal.lfilter` para los filtros FIR, la cual implementa la convolución directa del filtro sobre la señal. En el caso de los filtros IIR, se utilizó la función `scipy.signal.sosfiltfilt`, que aplica el filtro en ambas direcciones para evitar el desfase, lo cual equivale a un filtrado con fase lineal. Dado que el filtrado doble incrementa la pendiente de atenuación, los requerimientos originales de ripple y atenuación fueron predistorsionados para evitar sobredimensionar el filtro.    

```python
#%% APLICACIÓN DE LOS FILTROS Y COMPARACIÓN VISUAL

ecg_iir_cheby2 = sig.sosfiltfilt(mi_iir_1, ecg_one_lead)
ecg_iir_ellip = sig.sosfiltfilt(mi_iir_2, ecg_one_lead)

ecg_fir_firwin2 = sig.lfilter(mi_fir_1, [1], ecg_one_lead)
ecg_fir_firls = sig.lfilter(mi_fir_2, [1], ecg_one_lead)
```

Al comparar visualmente la señal original y la señal filtrada, se observa una clara mejora en la **suavidad de los trazos**, con una significativa **reducción del contenido de alta frecuencia**, que corresponde al ruido eléctrico o interferencias de mayor frecuencia. Esto es esperable dado que el filtro elimina componentes fuera de la banda pasante (1–35 Hz).

Además, se aprecia una mejora notable en el **nivel isoeléctrico**, que ahora se mantiene cercano a cero entre los latidos. Esto indica que las componentes de muy baja frecuencia, responsables del corrimiento del nivel de base, fueron correctamente atenuadas por el filtro.

En resumen, los filtros permiten preservar las características principales del ECG (como los complejos QRS, ondas P y T), al tiempo que eliminan componentes no deseadas que dificultan su análisis clínico.

A continuación se compara la señal filtrada por los distintos métodos e implementaciones en **regiones específicas de interés**, seleccionadas por dos motivos principales:

- Algunas de ellas presentan **interferencia significativa o ruido**, lo cual permite evaluar la capacidad de los filtros para eliminar contenido no deseado sin distorsionar la señal útil.
- Otras regiones fueron elegidas por estar **relativamente libres de ruido**, lo que permite analizar con claridad cómo cada implementación preserva la morfología de los componentes relevantes del ECG (ondas P, QRS, T).

Este enfoque dual permite realizar una evaluación cualitativa integral del desempeño de los filtros tanto en situaciones adversas como en condiciones ideales.
